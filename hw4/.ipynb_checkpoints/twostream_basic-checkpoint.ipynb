### Two-Stream Instability (PIC Method)

```python
data = nt2.Data("/dartfs/rc/lab/E/EPaCO/phys118/godfrey/hw2/twostream")
```

### Time Evolution of Total Electric Field Energy

```python
fdata = data.fields
```

```python
ind_out = np.arange(0, 500, 1)
E_energy = []

for i in ind_out:
    ex_array = np.array(fdata.isel(t=i).Ex.load())
    ez_array = np.array(fdata.isel(t=i).Ez.load())
    x_array  = np.array(fdata.isel(t=i).x.load())

    dx = x_array[1] - x_array[0]
    U_E = 0.5 * np.sum(ex_array**2 + ez_array**2) * dx
    E_energy.append(U_E)
```

```python
plt.figure(figsize=(7, 4))
plt.plot(ind_out, E_energy, label=r'$\frac{1}{2} \sum (E_{x}^2 + E_{z}^2) \Delta x$')
plt.xlabel(r'Time Step $(t/ \Delta t)$', fontsize=12)
plt.ylabel(r'Energy (dimensionless PIC units)', fontsize=12)
plt.title('Time Evolution of Total Electric Field Energy:   $u_{Stream}= \pm\ 0.1$', fontsize=12)
plt.legend(frameon=False, fontsize=15)
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.5)
plt.tight_layout()
#plt.savefig('E-field_energy_1.png')
```

## 2A) Convergence of PIC simulations:
For the initial problem setup of Exercise 2.2 $(u_{stream}= \pm\ 0.1)$, compute the total electric field energy
$$E(t) = \int |E|^2 \ dV$$

as a function of time. In the linear regime, $E(t) \approx exp(2\gamma t)$, where $\gamma$ is the growth rate. Fit an exponential to $E(t)$ during the linear growth phase to extract the numerical growth rate $\gamma_{num}$.

```python
# set linear regime
lin_min = 25
lin_max = 80

logE = np.log(E_energy[lin_min : lin_max]) # linearize

coeffs = np.polyfit(ind_out[lin_min : lin_max], logE, 1)
slope, intercept = coeffs

gamma = slope / 2 # growth rate

E_fitted = np.exp(intercept) * np.exp(slope * ind_out[lin_min : lin_max])
```

```python
plt.figure(figsize=(7, 4))

plt.plot(ind_out, E_energy, label=r'$E(t) = \int \left| E \right|^2 dV$')
plt.plot(ind_out[lin_min : lin_max], E_fitted, 'r--', label=rf'$E \approx exp(2\gamma t), \ \gamma={gamma:.5f}$') # fitted function

plt.xlim(0, 200) # only fitting to linear growth phase
plt.xlabel(r'Time Step $(t/ \Delta t)$', fontsize=12)
plt.ylabel(r'Energy (dimensionless PIC units)', fontsize=12)
plt.title('Time Evolution of Total Electric Field Energy (linear growth phase)', fontsize=12)
plt.legend(frameon=False, fontsize=15)
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.5)
plt.tight_layout()
plt.savefig('E-field_energy_fitted.png')
```
